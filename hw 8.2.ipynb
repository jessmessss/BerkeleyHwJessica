```python
import matplotlib.pyplot as plt 
import random

listA = [random.randint(0, 200) for _ in range(50)]
listB = [random.randint(0, 200) for _ in range(50)]
listC = [random.randint(0, 200) for _ in range(50)]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,12))
ax1.plot(listA, color='blue', linewidth=5, label='list A')
ax2.plot(listB, color='green', linestyle='dotted', label='list B')
ax1.set_title('Plot of list A and B')
ax1.legend()
ax1.grid(True)

x_values = list(range(50))
ax2.scatter(x_values, listC, color='purple', marker='^', label='list C')
ax2.set_title('plot fot list C')
ax2.legend
ax2.grid(True)

plt.tight_layout()
plt.show()
```
