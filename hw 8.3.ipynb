```python
import numpy as np 
import matplotlib.pyplot as plt 

def estimation_of_pi(N):
    x = np.random.uniform (0, 1, N)
    y = np.random.uniform (0, 1, N)

    #d=square root (x^2+y^2)
    distance = np.sqrt(x**2 + y**2)
    quarter_circle = np.sum(distance <=1)

    #1/4 *4 = full circle
    estimating_pi = (quarter_circle / N) * 4 
    return estimating_pi, x, y, distance

Ns= [10, 10**3, 10**4, 10**5, 10**6]
for N in Ns:
    estimating_pi, _, _, _ = estimation_of_pi(N)
    print(f"N = {N}, pi value = {estimating_pi}")

N =  10**4
estimating_pi, x, y, distance = estimation_of_pi(N)

#N values 

#radius 1 
x_values_in = x[distance <= 1]
x_values_out = x[distance > 1]
y_values_in = y[distance <= 1]
y_values_out = y[distance > 1]

plt.figure(figsize=(10,10))
plt.scatter(x_values_in,y_values_in, color='orange', s=1, label='Inner 1/4 of circle')
plt.scatter(x_values_out,y_values_out, color='pink', s=1, label='Outer 1/4 of circle')
plt.title(f"Monte Carlo -- calculation of PI")
plt.legend()
plt.axis('square')
plt.xlim(0,1)
plt.ylim(0,1)
plt.grid(True)
plt.show()
```
